## Roi

scr = img[0:200, 0:200] 是对图像 img 进行切片操作以提取ROI（感兴趣的区域）的操作。具体来说，这个操作从第0行到第199行和从第0列到第199列提取像素值，生成一个新的 NumPy 数组 ball，其中包含了原图像 img 中指定ROI的像素信息。这个操作可以用来对图像进行裁剪或者提取ROI区域进行进一步处理。

```python
import cv2

# 读取图像
img = cv2.imread("messi.jpg")

# 打印原始图像的形状
print("原始图像形状：", img.shape)

# 提取ROI
ball = img[0:200, 0:200]

# 打印ROI区域的形状
print("ROI区域形状：", ball.shape)

# 显示ROI
cv2.imshow("Ball", ball)

# 等待按键事件
cv2.waitKey(0)

# 关闭所有窗口
cv2.destroyAllWindows()
```

## 分离通道以及拼接图片

```python
import cv2 as cv

# 读取图像文件
img = cv.imread('opencv-logo-white.jpg')


print(img.shape)
# 分离三个颜色通道
img_b, img_g, img_r = cv.split(img)

# 将三个通道合并回一个图像
img2 = cv.merge((img_b, img_g, img_r))

print(img.shape)
# 显示图像
# 将三张图像沿着水平方向拼接
img_total = np.concatenate((img_b, img_g, img_r), axis=1)

cv.imshow('img_total', img_total)

# 等待按键事件
cv.waitKey(0)

# 释放窗口
cv.destroyAllWindows()
```

```python
import cv2 as cv

# 读取图像文件
img = cv.imread('opencv-logo-white.jpg')
img_b = img[:,:,0]
img_total = np.concatenate((img_b, img[:,:,0],img[:,:,1],img[:,:,2]), axis=1)
img[:,:,2]=0
cv.imshow('img', img)
cv.imshow('img_total', img_total)
# 等待按键事件
cv.waitKey(0)

# 释放窗口
cv.destroyAllWindows()
```

## 边界
1. CV.BORDER_REPLICATE：边框由图像的边缘像素复制而成
2. CV.BORDER_REFLECT：边框是图像边缘的镜像反射。
3. CV.BORDER_REFLECT_101：边框是图像边缘的镜像反射，但不包括角落像素。
4. CV.BORDER_WRAP：边框是图像的重复副本。
5. CV.BORDER_CONSTANT：边框被填充为一个常量值，该值由 value 参数指定。

```python
import cv2 as CV
import numpy as np
from matplotlib import pyplot as plt
BLUE = [255,0,0]
img1 = CV.imread('people.jpg')
replicate = CV.copyMakeBorder(img1,10,10,10,10,CV.BORDER_REPLICATE,)
reflect = CV.copyMakeBorder(img1,10,10,10,10,CV.BORDER_REFLECT)
reflect101 = CV.copyMakeBorder(img1,10,10,10,10,CV.BORDER_REFLECT_101)
wrap = CV.copyMakeBorder(img1,10,10,10,10,CV.BORDER_WRAP)
constant= CV.copyMakeBorder(img1,10,10,10,10,CV.BORDER_CONSTANT,value=BLUE)
plt.subplot(231),plt.imshow(img1,'gray'),plt.title('ORIGINAL')
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP')
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT')
plt.show()
```
