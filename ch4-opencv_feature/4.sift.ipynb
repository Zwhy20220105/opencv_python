## 尺度不变特征变换(SIFT)
1. 尺度空间极值检测
2. 关键点的精确定位
3. 方向分配
4. 关键点描述
5. 关键点匹配

应用图像拼接,比如全景照片的合成


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 定义高斯核的大小和标准差
ksize = (9, 9)
sigma = 2.0

# 生成二维离散高斯核
kernel = cv2.getGaussianKernel(ksize[0], sigma)  # 生成一维高斯核
kernel = np.outer(kernel, kernel)  # 进行外积操作，得到二维高斯核

# 显示高斯核图像
plt.imshow(kernel, cmap='jet')
plt.colorbar()
plt.title('Gaussian Kernel')
plt.show()
```

```python
import numpy as np
import cv2 as cv

img = cv.imread('../data/home.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

sift = cv.SIFT_create()
kp = sift.detect(gray, None)

img = cv.drawKeypoints(gray, kp, img)

cv.imshow('sift_keypoints.jpg', img)
cv.waitKey(0)
cv.destroyAllWindows()
```
