## ORB算法
- ORB算法是FAST关键点检测算法和BRIEF特征描述子的结合体，由于FAST算法的高效和BRIEF算法的优秀性能，使得ORB算法在保持有着和SIFT算法相近的性能的同时，具有更快的速度，因此ORB算法是一种非常好的特征提取算法。
- ORB算法的全称是Oriented FAST and Rotated BRIEF，即旋转的FAST关键点检测算法和旋转的BRIEF特征描述子的结合体。
- ORB算法的主要思想是在FAST关键点检测算法的基础上，对BRIEF特征描述子进行改进，使得BRIEF特征描述子具有旋转不变性，从而使得ORB算法具有旋转不变性。
- ORB算法的主要步骤如下：
    - 1.对输入图像进行高斯滤波，得到平滑后的图像；
    - 2.对平滑后的图像进行FAST关键点检测，得到关键点；
    - 3.对关键点进行亚像素精确化；
    - 4.对关键点进行方向分配；
    - 5.根据关键点的方向，对关键点周围的像素点进行BRIEF特征描述子的计算；
    - 6.对BRIEF特征描述子进行旋转，使得BRIEF特征描述子具有旋转不变性。
- ORB算法的主要优点是具有旋转不变性和尺度不变性，同时具有和SIFT算法相近的性能，但是ORB算法的主要缺点是对于视角变化和光照变化不具有不变性。
- ORB算法的主要应用是在SLAM（Simultaneous Localization and Mapping，即同时定位与地图构建）中，因为ORB算法具有旋转不变性和尺度不变性，因此ORB算法在SLAM中具有非常好的应用。
- ORB算法的主要优点是具有旋转不变性和尺度不变性，同时具有和SIFT算法相近的性能，但是ORB算法的主要缺点是对于视角变化和光照变化不具有不变性。


```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('../data/blox.jpg',0)
# 初始化ORB检测器
orb = cv.ORB_create()
# 用ORB寻找关键点
kp = orb.detect(img,None)
# 用ORB计算描述符
kp, des = orb.compute(img, kp)
# 仅绘制关键点的位置，而不绘制大小和方向
img2 = cv.drawKeypoints(img, kp, None, color=(0,255,0), flags=0)
plt.imshow(img2), plt.show()
```
