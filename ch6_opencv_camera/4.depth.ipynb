# 立体图的深度图

$$ {\notag} disparity = x - x' = \frac{Bf}{Z} $$
1. disparity: 视差
2. $x$和$x'$是图像平面上的点对应的 3D 场景点与相机投影中心之间的距离  
3. $B$是两个摄像机之间的距离(这是我们知道的)，$f$是相机的焦距(早已知晓)
4. $Z$是场景点的深度(我们需要计算的)


```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

imgL = cv.imread('../data/tsukuba_l.png',0)
imgR = cv.imread('../data/tsukuba_r.png',0)
#参数说明：
# numDisparities：最大视差值，必须是16的整数倍；
# blockSize：匹配的块大小，必须是大于等于1的奇数；
stereo = cv.StereoBM_create(numDisparities=16, blockSize=15)
disparity = stereo.compute(imgL,imgR)

# 得到的视差图,没有计算深度
# 如果需要计算深度的话，需要知道相机的内参和外参
plt.imshow(disparity,'gray')
plt.show()
```
