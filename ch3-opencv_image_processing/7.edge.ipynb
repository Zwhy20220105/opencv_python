## 边缘检测

1. 滤波减少噪声
2. 寻找图像的强度梯度
$$
Edge\_Gradient \; (G) = \sqrt{G_x^2 + G_y^2} \\ Angle \; (\theta) = \tan^{-1} \bigg(\frac{G_y}{G_x}\bigg)
$$
3. 非极大值抑制
4. 滞后阈值

```python
import cv2 as cv
import numpy as np

def update_canny(threshold1, threshold2):
    # 进行Canny边缘检测
    edges = cv.Canny(img, threshold1, threshold2)

    # 显示边缘图像
    cv.imshow('Canny Edge Detection', edges)
    cv.waitKey(1)

# 读取图像
img = cv.imread('../data/messi5.jpg', 0)

# 创建窗口
cv.namedWindow('Canny Edge Detection')

# 创建滑动条
cv.createTrackbar('Threshold 1', 'Canny Edge Detection', 0, 255, lambda x: update_canny(cv.getTrackbarPos('Threshold 1', 'Canny Edge Detection'), cv.getTrackbarPos('Threshold 2', 'Canny Edge Detection')))
cv.createTrackbar('Threshold 2', 'Canny Edge Detection', 0, 255, lambda x: update_canny(cv.getTrackbarPos('Threshold 1', 'Canny Edge Detection'), cv.getTrackbarPos('Threshold 2', 'Canny Edge Detection')))

# 显示原始图像
cv.imshow('Canny Edge Detection', img)
cv.waitKey(1)

# 进入主循环
while True:
    key = cv.waitKey(1)
    if key == ord('q'):
        break

cv.destroyAllWindows()
```

```python
import cv2 as cv
import threading

# 处理视频流的线程函数
def process_video():
    global frame
    while True:
        # 读取视频帧
        ret, frame = cap.read()
        if not ret:
            break

# 创建视频捕捉对象
cap = cv.VideoCapture(0)  # 0表示默认摄像头，也可以传入视频文件路径

# 初始化滑动条的初始值
threshold1 = 0
threshold2 = 0

# 创建窗口和滑动条
cv.namedWindow('Canny Edge Detection')
cv.createTrackbar('Threshold 1', 'Canny Edge Detection', 0, 255, lambda x: None)
cv.createTrackbar('Threshold 2', 'Canny Edge Detection', 0, 255, lambda x: None)

# 创建线程对象
thread = threading.Thread(target=process_video)

# 启动线程
thread.start()

# 等待一段时间确保线程已经开始处理视频流
cv.waitKey(300)

# 主线程中处理滑动条数值变化和显示结果
while True:

    # 获取滑动条数值
    threshold1 = cv.getTrackbarPos('Threshold 1', 'Canny Edge Detection')
    threshold2 = cv.getTrackbarPos('Threshold 2', 'Canny Edge Detection')

    # 复制当前帧用于处理
    current_frame = frame.copy()

    # 在主线程中进行Canny边缘检测
    edges = cv.Canny(current_frame, threshold1, threshold2)

    # 显示结果
    cv.imshow('Canny Edge Detection', edges)

    # 按下 'q' 键退出
    if cv.waitKey(1) == ord('q'):
        break

# 等待线程结束
#thread.join()

# 释放资源
cap.release()
cv.destroyAllWindows()
```

```python
import cv2 as cv
import numpy as np

# 打开摄像头
cap = cv.VideoCapture(0)
if not cap.isOpened():
    print("Cannot open camera")
    exit()
while True:
    # 获取每一帧图像
    ret, frame = cap.read()

    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    img = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    edges = cv.Canny(img, 100, 200)
    # 显示原始图像、掩膜和结果图像
    cv.imshow('edges', edges)
    
    # 按下Esc键退出程序
    k = cv.waitKey(5) & 0xFF
    if k == 27:
        break

# 关闭窗口
cv.destroyAllWindows()
```
